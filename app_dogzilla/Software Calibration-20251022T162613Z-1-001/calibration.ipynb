```python
# 导入库  import library
import ipywidgets.widgets as widgets
from IPython.display import display
```

```python
# Dogzilla drive library
from DOGZILLALib import DOGZILLA
g_dog = DOGZILLA()
```

# 机器狗软件舵机标定。
# 此功能重新校准舵机角度，出厂前以校准过，非必要可不进行校准。

```python
# 中文开关，默认为英文 Chinese switch. The default value is English
g_ENABLE_CHINESE = True

Name_widgets = {
    'Calibration': ("Calibration", "开始校准"),
    'Finish_Calib': ("Finish_Calib", "完成校准")
}
```

```python
# 开始校准 Calibration
button_Calibration = widgets.Button(        
    description=Name_widgets['Calibration'][g_ENABLE_CHINESE],        
    button_style='danger', # 'success', 'info', 'warning', 'danger' or ''     
    tooltip='Description',     
    icon='uncheck' )

# 完成校准 Finish_Calib
button_Finish_Calib = widgets.Button(        
    description=Name_widgets['Finish_Calib'][g_ENABLE_CHINESE],        
    button_style='success', # 'success', 'info', 'warning', 'danger' or ''     
    tooltip='Description',     
    icon='uncheck' )
```

```python
# 按键按下事件处理   Key press event processing
def on_button_clicked(b):
    b.icon = 'check'
    with output:
        print("Button clicked:", b.description)
    if b.description == Name_widgets['Calibration'][g_ENABLE_CHINESE]:
        g_dog.calibration(0x01)
    elif b.description == Name_widgets['Finish_Calib'][g_ENABLE_CHINESE]:
        g_dog.calibration(0x00)
        button_Calibration.icon = 'uncheck'
        button_Finish_Calib.icon = 'uncheck'
        
```

```python
# 关联按键事件回调 Button event callbacks
button_Calibration.on_click(on_button_clicked)
button_Finish_Calib.on_click(on_button_clicked)
```

## 布局控件并显示  Layout widgets and display them

```python
# 布局控件并显示  Layout widgets and display them
output = widgets.Output()
box_calib = widgets.HBox([button_Calibration, button_Finish_Calib])
box_display = widgets.VBox([box_calib, output])
display(box_display)
```
